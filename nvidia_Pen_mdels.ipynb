```python
## using nvidia qwen oopen source mdels

from langchain_nvidia_ai_endpoints import ChatNVIDIA


client = ChatNVIDIA(
  model="qwen/qwen3.5-122b-a10b",
  api_key= 0,  # use your own key here
  temperature=0.6,
  top_p=0.95,
  max_completion_tokens=16384,
)

lc_messages = [{"role":"user","content":"What is the recent status at the Strait of Hormuz? Give a brief summary."}]


for chunk in client.stream(lc_messages, chat_template_kwargs={"enable_thinking":True}):
  if chunk.additional_kwargs and "reasoning_content" in chunk.additional_kwargs:
    print(chunk.additional_kwargs["reasoning_content"], end="")
  print(chunk.content, end="")

```
