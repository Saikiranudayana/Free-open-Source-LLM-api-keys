```python
from openai import OpenAI

open_router_API = open_router_API1  # Use the variable from cell 1

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=open_router_API
)

response = client.chat.completions.create(
    model="openai/gpt-4o-mini",  # free & low token usage
    messages=[
        {"role": "system", "content": "You are a news reporter. Give 2 sentence summaries only."},
        {"role": "user", "content": "What is the latest news about AI?"}
    ],
    max_tokens=100  # limits response length
)
print(response.choices[0].message.content)
```
