```python
import os
from langchain_groq import ChatGroq
from langgraph.graph import StateGraph, START, END
from typing import TypedDict

os.environ["GROQ_API_KEY"] = Chat_groq_API

class State(TypedDict):
    question: str
    answer: str

llm = ChatGroq(model="openai/gpt-oss-120b")

def ask_groq(state: State) -> State:
    response = llm.invoke(state["question"])
    return {"question": state["question"], "answer": response.content}

graph = StateGraph(State)
graph.add_node("ask", ask_groq)
graph.add_edge(START, "ask")
graph.add_edge("ask", END)
app = graph.compile()

result = app.invoke({"question": "What is the recent status at the Strait of Hormuz? Give a brief summary.", "answer": ""})
print(result["answer"])
```
